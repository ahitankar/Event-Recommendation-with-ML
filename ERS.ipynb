```python
def StopwordsRemoval(data):   
    words = data.split()
   
    removeStopwords = [word for word in words if word.lower() not in stopwords.words('english')]
    
    return removeStopwords
```

```python
def priority(words):
    text = ' '.join(words)
    i = True
    domain_list = []
    for domain in domains['domain']:
        for item1 in Domain[domain]:
            if item1 in text.lower():
                domain_list.append(domain)
                i = False
    if i:
        domain_list.append('Other')
    event_list = []
    for event in events['event']:
        for item2 in Event[event]:
            if item2 in text.lower():
                event_list.append(event)
    return domain_list,event_list
```

```python
temp = []
dict = {}
for item in strings:
    words = StopwordsRemoval(item)
    pref = priority(words)
    t = []
    for D in list(set(pref[0])):
        if pref[1]:
            for E in list(set(pref[1])):
                row = list(df[(df['Domain']==D)&((df['Event1']==E)|(df['Event2']==E))]['Name'])
                t.append(row)
        else:
            row = list(df[(df['Domain']==D)]['Name'])
            dict[item]=row
    if t:
        temp_t = sum(t, [])
        dict[item]=list(set(temp_t))
```

```python
new_df = pd.DataFrame()
```

```python
for key, val in zip(list(dict.keys()),list(dict.values())):
    new_df = new_df.append({'Event' : key , 'Recommended Employees' : ', '.join(val)},ignore_index=True)
new_df = new_df[['Event','Recommended Employees']]
new_df.to_excel('result.xlsx',index=False)
```
